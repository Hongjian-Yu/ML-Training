```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy.optimize import minimize
```

```python
path = 'ex5data1.mat'
data = loadmat(path)
X = data['X']
Y = data['y']
Xtest = data['Xtest']
ytest = data['ytest']
Xval = data['Xval']
yval = data['yval']
```

```python
plt.scatter(X,Y)
# plt.show()
plt.scatter(Xtest,ytest)
# plt.show()
plt.scatter(Xval,yval)
# plt.show()
```

```python
X = np.insert(X,0,np.ones(X.shape[0]),axis=1)
Xtest = np.insert(Xtest,0,np.ones(Xtest.shape[0]),axis=1)
Xval = np.insert(Xval,0,np.ones(Xval.shape[0]),axis=1)
```

```python
def regularized_cost(theta,X,Y,lamda=1):
    
    y_p = X@theta - Y.reshape(-1)
    # print(X@theta)
    # print(Y)
    # print(y_p)
    temp = np.power((y_p),2) 
    temp_s = np.sum(temp) / (2 * X.shape[0])
    reg = np.power((theta[1:]),2)
    reg_s = np.sum(reg)  * lamda / (2 * X.shape[0])
    return reg_s + temp_s
```

```python
theta = np.ones(X.shape[1])
regularized_cost(theta,X,Y)
```

```python
def regularized_gradient(theta,X,Y,lamda=1):
    # print(X.shape)
    # print(theta.shape)
    # theta[0] = 0
    first = (X@theta - Y.reshape(-1))@X
    # print((X@theta).shape)
    # print(X.shape)
    # print(first.shape)
    second = np.multiply(theta,lamda)
    second[0] = 0
    return (first + second)/len(X)
```

```python
regularized_gradient(theta,X,Y,lamda=1)
```

```python
def training(X,Y,lamda):
    theta = np.ones(X.shape[1])
    result = minimize(fun=regularized_cost,x0=theta,args=(X,Y,lamda),method='TNC',jac=regularized_gradient,options={'disp':True})
    return result
```

```python
result = training(X,Y,1)
theta = np.array(result.x)
```

```python
fig,ax = plt.subplots(figsize=(10,6))
x = X[:,1:]
a = theta[1]
b = theta[0]
ax.scatter(x,Y.reshape(-1))
ax.plot(x,a*x+b)
plt.show()
```

```python

training_cost=[]
validation_cost=[]
m = X.shape[0]
for i in range(1,m+1):
    result = training(X[:i,:],Y[:i,:],0)
    training_cost.append(regularized_cost(result.x,X[:i,:],Y[:i,:],0))
    validation_cost.append(regularized_cost(result.x,Xval,yval,0))
    
```

```python
fig,ax = plt.subplots(figsize=(10,6))
ax.plot(np.arange(1,m+1),training_cost,label='training_cost')
ax.plot(np.arange(1,m+1),validation_cost,label='validation_cost')
plt.show()
```

```python
def get_more_feature(X,power):
    for i in range(2,power+1):
        X = np.insert(X,X.shape[1],np.power(X[:,1],i),axis=1)
    return X
```

```python
def get_mean_std(X):
    mean = np.mean(X,axis=0)
    std = np.std(X,axis=0)
    X[:,1:] = (X[:,1:] - mean[1:] ) / std[1:]
    # print(X.shape)
    return X
```

```python
power = 9
X = get_more_feature(X,power)
Xtest = get_more_feature(Xtest,power)
Xval = get_more_feature(Xval,power)

X = get_mean_std(X)
Xtest = get_mean_std(Xtest)
Xval = get_mean_std(Xval)
```

```python
def plot_learning_curve(X,Y,Xval,Yval,lamda):
    training_cost=[]
    validation_cost=[]
    m = X.shape[0]
    for i in range(1,m+1):
        result = training(X[:i,:],Y[:i,:],lamda)
        training_cost.append(regularized_cost(result.x,X[:i,:],Y[:i,:],lamda))
        validation_cost.append(regularized_cost(result.x,Xval,Yval,lamda))
    fig,ax = plt.subplots(figsize=(10,6))
    ax.plot(np.arange(1,X.shape[0]+1),training_cost)
    ax.plot(np.arange(1,X.shape[0]+1),validation_cost)
    plt.show()
    
   
```

```python
plot_learning_curve(X,Y,Xval,yval,lamda=0)
plot_learning_curve(X,Y,Xval,yval,lamda=1)
```

```python
plot_learning_curve(X,Y,Xval,yval,lamda=100)
```

```python
lamda_all = [0,0.001,0.003,0.01,0.03,0.1,0.3,1,3,10]
training_cost,validation_cost,test_cost = [], [], []
for i in lamda_all:
    result = training(X,Y,i)
    training_cost.append(regularized_cost(result.x,X,Y,0))
    validation_cost.append(regularized_cost(result.x,Xval,yval,0))
    test_cost.append(regularized_cost(result.x,Xtest,ytest,0))
fig,ax = plt.subplots(figsize=(10,6))
ax.plot(lamda_all,training_cost,c='blue')
ax.plot(lamda_all,validation_cost,c='orange')
ax.plot(lamda_all,test_cost,c='green')
plt.show()
```
