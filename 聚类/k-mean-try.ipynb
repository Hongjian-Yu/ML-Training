```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as ios
```

```python
path = 'ex7data2.mat'
data = ios.loadmat(path)
X = data['X']
print(data.keys())
```

```python
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(X[:,0],X[:,1])
plt.show()
```

```python
def find_centors(X,centors):
    idx = []
    for i in range(len(X)):
        idx.append(np.argmin(np.linalg.norm(X[i]-centors,axis=1)))
    return np.array(idx)
```

```python
init_centors = np.array([[3,3],[0,3],[4,2]])
idx = find_centors(X,init_centors)
plt.scatter(X[:,0],X[:,1],c=idx,cmap='rainbow')
plt.scatter(init_centors[:,0],init_centors[:,1],c=[0,1,2],s=100,marker='x',cmap='rainbow')
plt.show()
```

```python
def computer_centors(X,idx,k):
    new_centors=[]
    for i in range(k):
        new_centors.append(list(np.mean(X[idx==i],axis=0)))
    return np.array(new_centors)
```

```python
new_centors = computer_centors(X,idx,3)
plt.scatter(X[:,0],X[:,1],c=idx,cmap='rainbow')
plt.scatter(init_centors[:,0],init_centors[:,1],c=[0,1,2],s=100,marker='x',cmap='rainbow')
plt.scatter(new_centors[:,0],new_centors[:,1],c='k',s=200)
plt.show()
```

```python
def run_kmean(X,centors,iter):
    k=len(centors)
    centors_i = centors
    for i in range(iter):
        new_idx = find_centors(X,centors_i)
        centors_i = computer_centors(X,new_idx,k)
    return centors_i,new_idx
```

```python
last_centors,idx = run_kmean(X,init_centors,10)
plt.scatter(X[:,0],X[:,1],c=idx,cmap='rainbow')
plt.scatter(last_centors[:,0],last_centors[:,1],c=[0,1,2],marker='x',cmap='rainbow',s=200)
plt.show()
```
