```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as ios
```

```python
path = 'ex7data1.mat'
data = ios.loadmat(path)
X = data['X']
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
x_mean = X - np.mean(X,axis=0)
plt.scatter(x_mean[:,0],x_mean[:,1])
plt.show()
```

```python
Sigma = x_mean.T@x_mean / len(X)
print(Sigma.shape)
```

```python
U,S,V = np.linalg.svd(Sigma)
print(U)
```

```python
plt.scatter(x_mean[:,0],x_mean[:,1])
plt.plot([0,U[0,0]],[0,U[0,1]],c='r')
plt.plot([0,U[1,0]],[0,U[1,1]],c='k')
plt.show()
```

```python
U_reduce = U[:,0]
U_reduce.reshape((2,1))
z = x_mean@U_reduce
```

```python
plt.scatter(z[:],np.zeros(len(z)))
plt.show()
```

```python
print(z.shape)
print(U_reduce.shape)
z = z.reshape(50,1)
U_reduce = U_reduce.reshape(2,1)
X_approx = z@U_reduce.T
plt.scatter(x_mean[:,0],x_mean[:,1],c='b')
plt.scatter(X_approx[:,0],X_approx[:,1],c='r')
plt.show()
```
