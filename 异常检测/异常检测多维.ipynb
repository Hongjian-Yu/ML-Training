```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as ios
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
path = 'ex8data1.mat'
data = ios.loadmat(path)
print(data.keys())
X = data['X']
# X_val,X_test,Y_val,Y_test = train_test_split(data['Xval'],data['yval'],test_size=0.5)
X_val = data['Xval']
Y_val = data['yval']
```

```python
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(X[:,0],X[:,1])
plt.show()
```

```python
plt.scatter(X_val[:,0],X_val[:,1])

a = X_val[np.where(Y_val==1)[0]]
plt.scatter(a[:,0],a[:,1],c='r')
plt.show()
```

```python
mu = X.mean(axis=0)
sigma2 = np.mean(np.power((X-mu),2),axis=0)
print(mu)
print(sigma2)
```

```python
a = np.linspace(0,25,100)
b= np.linspace(0,25,100)
a, b = np.meshgrid(a,b)
Z = np.exp(-(np.power((a-mu[0]),2)/(2*sigma2[0])+np.power((b-mu[1]),2)/(2*sigma2[1])))
print(Z.shape)
```

```python
fig,ax = plt.subplots(figsize=(10,6))
ax.contour(a,b,Z,[10**-11,10**-7,10**-3,0.1])
ax.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from scipy import stats
```

```python
def select_epsilon(pval,yval):
    epsilon = np.linspace(pval.min(),pval.max(),10000)
    f1_best=[]
    
    for e in epsilon:
        y_p = (pval<e).astype('int')
        f1_best.append(f1_score(yval,y_p))

    f1_b = np.argmax(f1_best)
    return epsilon[f1_b],f1_best[f1_b]
```

```python
multi_normal = stats.multivariate_normal(mu,sigma2)
pval = multi_normal.pdf(X_val)
print(pval.shape)
```

```python
epsilon,f1 = select_epsilon(pval,Y_val)
print(epsilon)
print(f1)
```

```python
x_pdf = multi_normal.pdf(X)
ind = X[x_pdf<epsilon]
print(ind.shape)
plt.contour(a,b,Z,[epsilon])
plt.scatter(X[:,0],X[:,1])
plt.scatter(ind[:,0],ind[:,1],c='r')
plt.show()
```

```python
x_v_pdf = multi_normal.pdf(X_val)
plt.contour(a,b,Z,[epsilon])
plt.scatter(X_val[:,0],X_val[:,1])
plt.scatter(X_val[x_v_pdf<epsilon][:,0],X_val[x_v_pdf<epsilon][:,1],c='r')
plt.show()
```
