```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
path = 'ex6data1.mat'
data = loadmat(path)
X = data['X']
Y = data['y']
print(X.shape)
print(Y.shape)
```

```python
fig,ax = plt.subplots(figsize=(8,6))
X_true = []
X_false = []
y = Y.reshape(-1)
for i in range(len(y)):
    if y[i] == 0:
        X_false.append(list(X[i]))
    else:
        X_true.append(list(X[i]))
X_true = np.array(X_true)
X_false =  np.array(X_false)
ax.scatter(X_true[:,0],X_true[:,1],c='blue')
ax.scatter(X_false[:,0],X_false[:,1],c='red')
plt.show()
```

```python
from sklearn.svm import SVC
svc_1 = SVC(C=1,kernel='linear')
svc_1.fit(X,Y.reshape(-1))
svc_1.score(X,Y.reshape(-1))
```

```python
# a = np.linspace(X[:,0].min(),X[:,0].max(),500)
# b = np.linspace(X[:,1].min(),X[:,1].max(),500)
a = np.linspace(0,5,500)
b = np.linspace(1,5,500)

print(X[:,0].min())
print(X[:,0].max())
print(X[:,1].min())
print(X[:,1].max())
aa,bb = np.meshgrid(a,b)
z = np.c_[aa.reshape(-1),bb.reshape(-1)]
out = svc_1.predict(z)
out = out.reshape(aa.shape)
fig,ax = plt.subplots(figsize=(10,6))
ax.contour(aa,bb,out)
ax.scatter(X_true[:,0],X_true[:,1],c='blue')
ax.scatter(X_false[:,0],X_false[:,1],c='red')
plt.show()
```
