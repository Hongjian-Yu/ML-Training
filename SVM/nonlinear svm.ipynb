```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
path = 'ex6data2.mat'
data = sio.loadmat(path)
X = data['X']
Y = data['y']
print(data.keys())
```

```python
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(X[:,0],X[:,1],c=Y.flatten(),cmap='jet')
plt.show()
```

```python
from sklearn.svm import SVC
svc_1 = SVC(C=1,kernel='rbf',gamma=1000)
svc_1.fit(X,Y.flatten())
svc_1.score(X,Y.flatten())
```

```python
a = np.linspace(0,1.2,500)
b = np.linspace(0.3,1,500)
aa,bb = np.meshgrid(a,b)
z = svc_1.predict(np.c_[aa.flatten(),bb.flatten()])
out = z.reshape(aa.shape)
plt.contour(aa,bb,out)
plt.scatter(X[:,0],X[:,1],c=Y.flatten(),cmap='jet')
plt.show()
```
