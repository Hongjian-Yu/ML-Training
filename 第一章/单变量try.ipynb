```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'ex1data1.txt'
data = pd.read_csv(path,names=['x','y'])
```

```python
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(data['x'],data['y'])
plt.show()
```

```python
data.insert(0,'b',1)
```

```python
X = data.iloc[:,0:2].values
Y = data.iloc[:,2:].values
w = np.matrix([0,0])
w = np.float64(w)
```

```python
def cost(X,Y,w):
    tem = np.power((X@w.T-Y),2)
    return tem.sum() / (2*len(X))
```

```python
def gradient_descent(X,Y,w,lr,iter):
    
    for i in range(iter):

        error = X@w.T - Y

        tem = X.T@error

        w -= tem.T * lr / len(X)

    return w
```

```python
w = gradient_descent(X,Y,w,0.02,1000)
fig,ax = plt.subplots(figsize=(10,6))
X = np.linspace(data['x'].min(),data['x'].max())
print(X)
Y = w[0,1] * X + w[0,0]
ax.plot(X,Y)
ax.scatter(data['x'],data['y'])
plt.show()
```
