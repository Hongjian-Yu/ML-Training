```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'ex1data1.txt'
```

```python
with open(path,'r') as file:
    file_all = file.readlines()
x = list()
y = list()
for i in file_all:
    a = i.split(',')
    x.append(float(a[0]))
    y.append(float(a[1]))
```

```python
xplot = x
yplot = y
plt.scatter(xplot,yplot)
plt.show()
```

```python
class Training:
    def __init__(self,x,y):
        self.w = 0
        self.b = 0
        self.x = x
        self.y = y

    def train(self):
        for epoch in range(10000):
            temp_1_list = list()
            temp_list=list()
            for i in range(len(self.x)):
                y_pred = self.x[i] * self.w + self.b
                temp = y_pred - self.y[i]
                temp_1 = temp * self.x[i]
                temp_1_list.append(temp_1)
                temp_list.append(temp)
            a = sum(temp_1_list) / len(self.x)
            b = sum(temp_list) / len(self.x)
            self.w = self.w - a * 0.0001
            self.b = self.b - b * 0.0001
            
                
```

```python
hun = Training(x,y)
hun.train()
print(hun.w)
print(hun.b)
```

```python
plt.scatter(xplot,yplot)
a = [5,22]
b = [5*hun.w+hun.b,22*hun.w+hun.b]
plt.plot(a,b)
plt.show()
```
