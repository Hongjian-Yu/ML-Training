```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = 'ex1data1.txt'
data = pd.read_csv(path,names=['x','y'])
data.insert(0,'b',1)
data.head()
```

```python
col = data.shape[1]
X = data.iloc[:,:col-1]
Y = data.iloc[:,col-1:]
X = np.matrix(X.values)
Y = np.matrix(Y.values)
w = np.matrix([0,0])
plt.scatter(data['x'],data['y'])
plt.show()
```

```python
def cost(X,Y,w):
    temp = np.power((X.dot(w.T) - Y),2)
    
    return np.sum(temp) / (2*len(X))
```

```python
def normal(X,Y):
    w = np.linalg.inv(X.T@X)@X.T@Y
    return w
```

```python
w = normal(X,Y)
w[1,0]
```

```python
plt.figure(figsize=(12,8))
a = [data.x.min(),data.x.max()]
b = [w[0,0] + w[1,0] * data.x.min() , w[0,0] + w[1,0] * data.x.max() ]
plt.plot(a,b)
plt.scatter(data['x'],data['y'])
plt.show()
```
