```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
from scipy.io import loadmat
```

```python
path = 'ex3data1.mat'
data = loadmat(path)
X = data['X']
Y = data['y']
Y_else = Y
print(X.shape)
print(Y.shape)
print(Y)
```

```python
def plot_an_image(image):
    fig,ax = plt.subplots(figsize=(1,1))
    ax.matshow(image.reshape(20,20),cmap=matplotlib.cm.binary)
    plt.xticks(np.array([]))
    plt.yticks(np.array([]))
    plt.show()
pick_one = np.random.randint(0, 5000)
plot_an_image(X[pick_one, :])
print('this should be {}'.format(Y[pick_one]))
print(Y)
```

```python
X = np.insert(X,0,np.ones(X.shape[0]),axis=1)
w = np.matrix(np.random.rand(X.shape[1]))
w = np.float64(w)
print(Y)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def lost(X,Y,w):
    first = np.multiply(Y, np.log(sigmoid(X.dot(w.T))))
    second = np.multiply((1 - Y) ,np.log(1 - sigmoid(X.dot(w.T))))
    return -np.mean(first+second)
```

```python
def gradientDescent(X,Y,w,lr,iter):
    
    cost = np.zeros(iter)
    for i in range(iter):

        error = sigmoid(X.dot(w.T)) - Y

        temp = X.T.dot(error)

        w = w - lr / len(X) * temp.reshape(1,-1)
    
        cost[i] = lost(X,Y,w)
    
    return w,cost
```

```python
X_1 = X.copy()
Y_1 = Y.copy()
w_all = list()

for i in range(1,11):
    
    x = X_1.copy()
    y = Y_1.copy()
    print(Y)
    for j in y:
        if j[0] == i:
            j[0] = 1
        else:
            j[0] = 0
    print(Y)
    w,cost = gradientDescent(x,y,w,0.01,1000)
    b = np.arange(1000)
    plt.plot(b,cost)
    plt.show()
    
    w_all.append(w.tolist())
    w = np.matrix(np.random.rand(X.shape[1]))
    w = np.float64(w)
```

```python
w_a = np.array(w_all)
w_a = w_a.reshape(10,401)
print(w_a.shape)
print(Y)
```

```python
t = X.dot(w_a.T)
li = np.argmax(t,axis=1)
li = li + 1
print(li)
print(Y)
Y_e = Y_else.reshape(-1)
print(Y)
```

```python
cnt=0
for i in range(len(Y_e)):
    if Y_e[i] == li[i]:
        cnt += 1
print(cnt/len(Y))
print(cnt)
```
